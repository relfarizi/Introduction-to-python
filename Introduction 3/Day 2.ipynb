Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note : warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

Expected Result :



```python
import matplotlib.pyplot as plt
grafik = [cos,cos3,sin,sin2]
color = ["green","blue","red","yellow"]
line = ["solid","--","-.","None"]
marker = ["None","None","None","x"]
legend = ["cos x","3 cos x","sin x-90","sinx2"]
fig, ax = plt.subplots()
for i in range(len(grafik)):
  ax.plot(grafik[i],color=color[i],marker=marker[i],linestyle = line[i],label=legend[i])
ax.set_xlabel("X-Axis Label")
ax.set_ylabel("Y-Axis Label")
ax.set_title("Grafik Cos Sin")
ax.legend(fancybox=True, framealpha=0.5,loc="center")

plt.show()
```
